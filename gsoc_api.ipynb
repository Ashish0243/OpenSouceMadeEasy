```python
import httpx
```

```python
class GsocApiClient:
    def __init__(self):
        self.base_url = "https://api.gsocorganizations.dev"
    
    # def _headers(self):
    #     return{
    #         "Accept": "application/json",
    #     }

    def get_all_organizations(self):
        response = httpx.get(f"{self.base_url}/2025.json")
        response.raise_for_status()
        return response.json()
```

```python
from dotenv import load_dotenv
```

```python
import os
load_dotenv()
api_key=os.getenv('GEMINI_API_KEY')
```

```python
client=GsocApiClient()
ans=client.get_all_organizations()
print(ans['organizations'][0:2])
```
